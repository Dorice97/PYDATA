Kernel Density Estimate (KDE PLOTS)
-visualize estimated probability density of a function(PDF)
-continious and non parametric data


```python
temp = pd.Series([1, 2, 2.5, 3, 3.5, 4, 5])
display = temp.plot.kde()
```
